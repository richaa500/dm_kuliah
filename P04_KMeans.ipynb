```python
import pandas as pd
import numpy as np

df = pd.DataFrame(
    {"panjang": [1.4, 1.5, 1.4, 4.7, 3.9, 5.0, 5.1, 6.7, 5.7, 6.1],
     "lebar": [0.2, 0.2, 0.3, 1.4, 1.1, 1.7, 2.4, 2.2, 2.1, 2.3]}
)
df
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
df1 = df[["panjang", "lebar"]]
X = df1[["panjang", "lebar"]]
X
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x = 'panjang', y = 'lebar', data=X, s = 60)
plt.xlabel('panjang')
plt.ylabel('lebar')
plt.title('K-Means')
plt.show()
```

```python
from sklearn.cluster import KMeans
km1 = KMeans(n_clusters=3)
km1.fit(X)
y = km1.predict(X)
```

```python
df1["label"] = y
df1
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x = 'panjang', y = 'lebar', hue = "label",
                palette = ['green', 'orange', 'brown'], legend='full', data = df1, s = 60)

plt.scatter(km1.cluster_centers_[:, 0], km1.cluster_centers_[:, 1], s = 60, c = 'black', label = 'Centroids')
plt.xlabel('panjang')
plt.ylabel('lebar')
plt.title('manual')
plt.show()
```
